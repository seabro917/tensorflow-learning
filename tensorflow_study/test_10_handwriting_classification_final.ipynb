```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

```python
# Load the sample data
mnist = input_data.read_data_sets("MNIST_data",one_hot = True)
# Set the size of batch
batch_size = 100
# Number of batches in total
number_of_batch = mnist.train.num_examples // batch_size


# Define four place holders (Each picture is of size 28*28, which is 784, and will be extended as a one-dimensional vector)
x = tf.placeholder(tf.float32,[None, 784])
# Ten different kinds of numbers in total, ranging from 0 to 9.
y = tf.placeholder(tf.float32,[None, 10])
# How many neurons work during each epoch 
keep_prob = tf.placeholder(tf.float32)
learning_rate = tf.Variable(0.001,dtype=tf.float32)

# Two hidden layer (Mention the way of initializing the weights and biases)
# 500 neurons
Weight_L1 = tf.Variable(tf.truncated_normal([784,500],stddev = 0.1))
Biase_L1 = tf.Variable(tf.zeros([500]) + 0.1)
L1 = tf.nn.tanh(tf.matmul(x,Weight_L1) + Biase_L1)
L1_drop = tf.nn.dropout(L1, keep_prob) # The proportion of the working neurons.
# 300 neurons
Weight_L2 = tf.Variable(tf.truncated_normal([500,300],stddev = 0.1))
Biase_L2 = tf.Variable(tf.zeros([300]) + 0.1)
L2 = tf.nn.tanh(tf.matmul(L1_drop,Weight_L2) + Biase_L2)
L2_drop = tf.nn.dropout(L2, keep_prob) # The proportion of the working neurons.


# Output layer 10 neurons (One_hot)
Weight_L3 = tf.Variable(tf.truncated_normal([300,10],stddev = 0.1))
Biase_L3 = tf.Variable(tf.zeros([10]) + 0.1)
prediction = tf.nn.softmax(tf.matmul(L2_drop,Weight_L3) + Biase_L3)


# Cross-entropy cost function
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=tf.matmul(L2_drop,Weight_L3) + Biase_L3))
train_step = tf.train.AdamOptimizer(learning_rate).minimize(loss)

# Evaluate the accuracy of the model (argmax return the index of the max value)
# correction_prediction is a list of type boolean
correct_prediction = tf.equal(tf.argmax(y,1),tf.argmax(prediction,1)) # One-dimensional tensor
# Calculate the accuracy (Transfrom the type into float32 format)
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    # 21 cycles
    for epoch in range(21):
        sess.run(tf.assign(learning_rate,0.001 * (0.95 ** epoch)))
        # Iteration through all images
        for batch in range(number_of_batch):
            # Get the sample and corresponding label
            batch_sample,batch_label = mnist.train.next_batch(batch_size)
            sess.run(train_step,feed_dict = {x:batch_sample,y:batch_label,keep_prob:1.0})
        lr = sess.run(learning_rate)
        # Calculate accuracy after one iteration is done
        
        # Accuracy for testing set
        test_acc = sess.run(accuracy,feed_dict = {x:mnist.test.images,y:mnist.test.labels,keep_prob:1.0})
        # Accuracy for training set
        train_acc = sess.run(accuracy,feed_dict = {x:mnist.train.images,y:mnist.train.labels,keep_prob:1.0})
        print("Iter: " + str(epoch) + ", Testing accuracy: " + str(test_acc) + ", Training accuracy: " + str(train_acc) + ",Learning rate:" + str(lr))
        
```
